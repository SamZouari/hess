# Création d'une image 3D

## Définition de la géométrie

```python
# Target definition
target_position = SkyCoord(359.94, -00.04, frame="galactic", unit="deg")

# Define geoms
emin, emax = [0.5, 100] * u.TeV
energy_axis = MapAxis.from_bounds(
    emin.value, emax.value, 10, unit="TeV", name="energy", interp="log"
)
geom = WcsGeom.create(
    skydir=target_position,
    binsz=0.04,
    width=(2, 2),
    coordsys="CEL",
    proj="CAR",
    axes=[energy_axis],
)

etrue_axis = MapAxis.from_bounds(
    0.1, 200, 20, unit="TeV", name="energy", interp="log"
)

geom_true = WcsGeom.create(
    skydir=target_position,
    binsz=0.04,
    width=(2, 2),
    coordsys="CEL",
    proj="CAR",
    axes=[etrue_axis],
)
```

```python
# Define the source model - Use a pointsource + integrated power law model to directly get flux

spatial_model = PointSpatialModel(
    lon_0=target_position.l, lat_0=target_position.b, frame="galactic"
)

spectral_model = PowerLawSpectralModel(
    index=2.3,
    amplitude=2.0e-11 * u.Unit("1 / (cm2 s TeV)"),
    reference=1 * u.TeV,
)

spectral_model.parameters["index"].frozen = False

sky_model = SkyModel(
    spatial_model=spatial_model, spectral_model=spectral_model, name="SgrA*"
)

sky_model.parameters["lon_0"].frozen = True
sky_model.parameters["lat_0"].frozen = True
```

```python
# psf_kernel and MapMaker for each segment, avec méthode basique pour l'estimation du fond

def make_map_dataset(
    observations, target_pos, geom, geom_true, offset_max=2 * u.deg
):
    maker = MapMaker(geom, offset_max, geom_true=geom_true)
    maps = maker.run(observations)
    table_psf = make_mean_psf(observations, target_pos)

    # PSF kernel used for the model convolution
    psf_kernel = PSFKernel.from_table_psf(
        table_psf, geom_true, max_radius="0.3 deg"
    )
    edisp = make_mean_edisp(
        observations,
        target_pos,
        e_true=geom_true.axes[0].edges,
        e_reco=geom.axes[0].edges,
    )
    
    # méthode basique
    background_model = BackgroundModel(maps["background"])
    background_model.parameters["norm"].frozen = False
    background_model.parameters["tilt"].frozen = True

    dataset = MapDataset(
        counts=maps["counts"],
        exposure=maps["exposure"],
        background_model=background_model,
        psf=psf_kernel,
        edisp=edisp,
    )
    return dataset
```

```python
# psf_kernel and MapMaker for each segment, avec méthode de l'anneau pour l'estimation du fond

def make_map_dataset_ring(
    observations, target_pos, geom, geom_true, offset_max=2 * u.deg
):
    
    # souci là avec le MapMaker, il faut MapMakerRing, qui ne marche pas, car sinon on n'a pas les bonnes cartes en sortie de maker.run (sauf si on utilise run_images qui n'est pas pratique)
    maker = MapMaker(geom, offset_max, geom_true=geom_true)
    
    
    maps = maker.run(observations)
    table_psf = make_mean_psf(observations, target_pos)

    # PSF kernel used for the model convolution
    psf_kernel = PSFKernel.from_table_psf(
        table_psf, geom_true, max_radius="0.3 deg"
    )
    edisp = make_mean_edisp(
        observations,
        target_pos,
        e_true=geom_true.axes[0].edges,
        e_reco=geom.axes[0].edges,
    )
    
    # méthode Ring Background
    
    
    
    
    background_model = BackgroundModel((maps["exposure_on"]/maps["exposure_off"]) * maps["off"], norm=1.1, tilt=0.0)
    background_model.parameters["norm"].frozen = False
    background_model.parameters["tilt"].frozen = True

    dataset = MapDataset(
        counts=maps["counts"],
        exposure=maps["exposure"],
        background_model=background_model,
        psf=psf_kernel,
        edisp=edisp,
    )
    return dataset
```

```python
%%time

# pas adapté ici, ça prend trop de temps et la plupart des time_interval sont vides (non en fait car on les a définit nous-mêmes)

datasets = []

for time_interval,obs in zip(year_intervals,obs_set):
    # get filtered observation lists in time interval
    #obs = observations.select_time(time_interval)
    # Proceed with further analysis only if there are observations
    # in the selected time window
    #if len(obs) == 0:
    #    log.warning(
    #        "No observations found in time interval:"
    #        "{t_min} - {t_max}".format(
    #            t_min=time_interval[0], t_max=time_interval[1]
    #        )
    #    )
    #    continue
    dataset = make_map_dataset(obs, target_position, geom, geom_true)
    dataset.counts.meta["t_start"] = time_interval[0]
    dataset.counts.meta["t_stop"] = time_interval[1]
    datasets.append(dataset)
```

### Ajustement d'un modèle

```python
from astropy.modeling import models, fitting
```

```python
lc.compute_chisq()
```

```python
def chisq(obs, exp):
    chisq = 0
    for i in range(len(obs)):
        chisq = chisq + ((obs[i]-exp)**2)/(exp**2)
    return chisq


def chisqr(obs, exp, error):
    chisqr = 0
    for i in range(len(obs)):
        chisqr = chisqr + ((obs[i]-exp)**2)/(error[i]**2)
    return chisqr
```

```python
import scipy.stats as stats

A1 = lc.table["flux"].data
A2 = np.mean(A1)
err_bars = lc.table["flux_err"].data
chi2 , pval = chi2, pval = stats.chisquare(A1, A2)
print(chi2, pval, chisqr(A1, A2, err_bars))
```

```python


# Generate data
x = (lc.table["time_min"].data + lc.table["time_max"].data)/2
y = A1
y_uncs = lc.table["flux_err"].data

x = np.delete(x, 12)
y = np.delete(y, 12)
y_uncs = np.delete(y_uncs, 12)


# Fit the data using a box model.
# Bounds are not really needed but included here to demonstrate usage.
t_init = models.Const1D(1e-12)
fit_t = fitting.LevMarLSQFitter()
t = fit_t(t_init, x, y, weights=1.0/y_uncs)


# Plot the data with the best-fit model
plt.figure(figsize=(8,5))
plt.plot(x, y, 'ko')
plt.errorbar(x, y, yerr=y_uncs, fmt='kx', label="data")
plt.plot(x, t(x), label='constant')
plt.xlabel('time')
plt.ylabel('Flux')
plt.legend(loc=2)
```

```python
valfit = t(55000)

A1 = lc.table["flux"].data
A2 = valfit
err_bars = lc.table["flux_err"].data
chi2 , pval = chi2, pval = stats.chisquare(A1, A2)
print(chi2, pval, chisqr(A1, A2, err_bars))
```

# Courbes de lumière en 1D

```python
from regions import CircleSkyRegion
from astropy.coordinates import Angle
from gammapy.spectrum import (
    SpectrumExtraction,
    ReflectedRegionsBackgroundEstimator,
    )
```

```python
# Target definition
on_region_radius = Angle("0.12 deg")  # paramètre qui change sensiblement le flux observé, donc il y a une "bonne" valeur probablement
on_region = CircleSkyRegion(center=target_position, radius=on_region_radius)
```

```python
from gammapy.modeling import Datasets

# à noter qu'ici on applique la technique dite "région mirroir" pour l'estimation du fond, donc pas forcément la meilleure, pour l'anneau voire la classe RingBackgroundEstimator

def dataset_maker_1D(obs):
    bkg_estimator = ReflectedRegionsBackgroundEstimator(
    on_region=on_region, observations= obs
    )
    bkg_estimator.run()
    
    extraction = SpectrumExtraction(
    observations=obs,
    bkg_estimate=bkg_estimator.result,
    containment_correction=True,
    e_reco=energy_axis.edges,
    e_true=etrue_axis.edges,
    )
    
    extraction.run()
    datasets_1d = Datasets(extraction.spectrum_observations).stack_reduce() 
    
    return datasets_1d
```

### On essaie de faire les datasets 1D avec RingBackground

```python


#images = {
#            'counts': Map.read(filename, hdu='counts'),
#            'exposure_on': Map.read(filename, hdu='exposure'),
#            'exclusion': Map.read(filename, hdu='exclusion'),
#        }

def dataset_maker_1D_ring(images):   
    bkg_estimator = RingBackgroundEstimator(
        r_in='0.35 deg', width='0.3 deg'
    )
    bkg_estimator.run(images)
    
    extraction = SpectrumExtraction(
    observations=obs,
    bkg_estimate=bkg_estimator.result,
    containment_correction=True,
    e_reco=energy_axis.edges,
    e_true=etrue_axis.edges,
    )
    
    extraction.run()
    datasets_1d = Datasets(extraction.spectrum_observations).stack_reduce() # renvoie une liste avec toutes les obs de l'année
    
    return datasets_1d
```

```python
%%time

datasets_1D = []

for time_interval,obs in zip(year_intervals, obs_set) :
    dataset = dataset_maker_1D(obs)
    dataset.counts.meta = dict()
    dataset.counts.meta["t_start"] = time_interval[0]
    dataset.counts.meta["t_stop"] = time_interval[1]
    datasets_1D.append(dataset)
```

```python
for dataset in datasets_1D:
    # Copy the source model
    model = spectral_model.copy()
    model.name = "SgrA*"
    dataset.model = model
```

```python
lc_maker_1d = LightCurveEstimator(datasets_1D, source="SgrA*", reoptimize=False)
```

```python
%%time
lc_1d = lc_maker_1d.run(e_ref=1 * u.TeV, e_min=1.0 * u.TeV, e_max=10.0 * u.TeV)
```

```python
plt.figure(figsize=(8, 6))
ax = lc_1d.plot(marker="o", label="1D")
lc.plot(ax=ax, marker="o", label="3D")
plt.legend()
plt.savefig("light_curves1D3D.pdf")
```
