# Analyse 1D des 36 dernières observations de Sgr A* avec HESS, en suivant le tuto spectrum_analysis de la version 0.14

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Check package versions
import gammapy
import numpy as np
import astropy
import regions

print("gammapy:", gammapy.__version__)
print("numpy:", np.__version__)
print("astropy", astropy.__version__)
print("regions", regions.__version__)
```

```python
import astropy.units as u
from astropy.coordinates import SkyCoord, Angle
from regions import CircleSkyRegion
from gammapy.maps import Map
from gammapy.modeling import Fit, Datasets
from gammapy.data import ObservationStats, ObservationSummary, DataStore
from gammapy.modeling.models import (
    PowerLawSpectralModel,
    create_crab_spectral_model,
)
from gammapy.spectrum import (
    SpectrumExtraction,
    FluxPointsEstimator,
    FluxPointsDataset,
    ReflectedRegionsBackgroundEstimator,
)
```

```python
datastore = DataStore.from_dir("../data/ash_stereo_Prod17_Calib0834_thsq64")
# on utilise la liste faite dans l'analyse 3D
#obs_ids = [140950, 146965, 146966, 146967, 146968, 146969, 146970, 147007, 147008, 147009, 147010, 147011, 147056, 147121, 147805, 147806, 147807, 147808, 147809, 147810, 147843, 147844, 147845, 147846, 147847, 147848, 147849, 147876, 147877, 147878, 147879, 150146]
#obs_ids2 = [147877, 147878, 147879, 150146]
```

```python
from astropy.coordinates import Angle

selection = dict(type='sky_circle', frame='galactic',
                 lon=Angle(0, 'deg'),
                 lat=Angle(0, 'deg'),
                 radius=Angle(2, 'deg'),
                 border=Angle(0, 'deg'))

selected_obs_table = datastore.obs_table.select_observations(selection)
```

```python
ids_gc = list(selected_obs_table["OBS_ID"])
obs_ids = ids_gc[:300]
```

```python
observations = datastore.get_observations(obs_ids)
```

```python
for obs in observations:
    try:
        obs.aeff
        obs.edisp
    except:
        obs_ids.remove(obs.obs_id)
        print("Observation retirée : " + str(obs.obs_id))

observations = datastore.get_observations(obs_ids)
```

```python
target_position = SkyCoord(0.0, 0.0, unit="deg", frame="galactic")
on_region_radius = Angle("0.10 deg")
on_region = CircleSkyRegion(center=target_position, radius=on_region_radius)
```

```python
exclusion_region = CircleSkyRegion(
    center=SkyCoord(0, 0, unit="deg", frame="galactic"),
    radius=0.5 * u.deg,
)

skydir = target_position.galactic
exclusion_mask = Map.create(
    npix=(100, 100), binsz=0.05, skydir=skydir, proj="TAN", coordsys="GAL"
)

mask = exclusion_mask.geom.region_mask([exclusion_region], inside=False)
exclusion_mask.data = mask
exclusion_mask.plot();
```

```python
background_estimator = ReflectedRegionsBackgroundEstimator(
    observations=observations,
    on_region=on_region,
    exclusion_mask=exclusion_mask,
)

background_estimator.run()
```

```python
stats = []
for obs, bkg in zip(observations, background_estimator.result):
    stats.append(ObservationStats.from_observation(obs, bkg))

print(stats[1])

obs_summary = ObservationSummary(stats)
fig = plt.figure(figsize=(10, 6))
ax1 = fig.add_subplot(121)

obs_summary.plot_excess_vs_livetime(ax=ax1)
ax2 = fig.add_subplot(122)
obs_summary.plot_significance_vs_livetime(ax=ax2);
```

```python
e_reco = np.logspace(-1, np.log10(100), 20) * u.TeV
e_true = np.logspace(np.log10(0.05), 2.3, 100) * u.TeV
```

```python
extraction = SpectrumExtraction(
    observations=observations,
    bkg_estimate=background_estimator.result,
    containment_correction=False,
    e_reco=e_reco,
    e_true=e_true,
)
```

```python
%%time
extraction.run()
```

```python
# Add a condition on correct energy range in case it is not set by default
extraction.compute_energy_threshold(method_lo="area_max", area_percent_lo=10.0)
```

```python
# Requires IPython widgets
# extraction.spectrum_observations.peek()
from gammapy.modeling import Datasets
datasets_stacked = Datasets(extraction.spectrum_observations).stack_reduce()
datasets_stacked.peek()
```

## On prépare les données pour l'ajustement

```python
datasets_joint = extraction.spectrum_observations

datasets_stacked = Datasets(datasets_joint).stack_reduce()

e_min, e_max = 0.3, 30
e_edges = np.logspace(np.log10(e_min), np.log10(e_max), 6) * u.TeV
```

## On teste le modèle d'une loi de puissance

```python
model = PowerLawSpectralModel(
    index=2, amplitude=2e-11 * u.Unit("cm-2 s-1 TeV-1"), reference=1 * u.TeV
)

for dataset in datasets_joint:
    dataset.model = model

fit_joint = Fit(datasets_joint)
result_joint = fit_joint.run()

# we make a copy here to compare it later
model_best_joint = model.copy()
model_best_joint.parameters.covariance = result_joint.parameters.covariance
```

```python
print(result_joint)
```

```python
plt.figure(figsize=(8, 6))
ax_spectrum, ax_residual = datasets_joint[20].plot_fit()
ax_spectrum.set_ylim(0, 25)
```

```python
fpe = FluxPointsEstimator(datasets=datasets_joint, e_edges=e_edges)
flux_points = fpe.run()
```

```python
flux_points.table_formatted
```

```python
plt.figure(figsize=(8, 5))
flux_points.table["is_ul"] = flux_points.table["ts"] < 4
ax = flux_points.plot(
    energy_power=2, flux_unit="erg-1 cm-2 s-1", color="darkorange"
)
flux_points.to_sed_type("e2dnde").plot_likelihood(ax=ax)
```

```python
flux_points_dataset = FluxPointsDataset(
    data=flux_points, model=model_best_joint
)
```

```python
plt.figure(figsize=(8, 6))
flux_points_dataset.peek();
```

## On stack les observations

```python
model = PowerLawSpectralModel(
    index=2, amplitude=2e-11 * u.Unit("cm-2 s-1 TeV-1"), reference=1 * u.TeV
)

datasets_joint = extraction.spectrum_observations

datasets_stacked = Datasets(datasets_joint).stack_reduce()
```

```python
datasets_stacked.model = model
stacked_fit = Fit([datasets_stacked])
result_stacked = stacked_fit.run()

# make a copy to compare later
model_best_stacked = model.copy()
model_best_stacked.parameters.covariance = result_stacked.parameters.covariance
```

```python
print(result_stacked)
```

```python
plt.figure(figsize=(8, 6))
ax_spectrum, ax_residual = datasets_stacked.plot_fit()
#ax_spectrum.set_ylim(0, 25)
```

```python
fpe = FluxPointsEstimator(datasets=datasets_stacked, e_edges=e_edges)
flux_points = fpe.run()
```

```python
plt.figure(figsize=(8, 5))
flux_points.table["is_ul"] = flux_points.table["ts"] < 4
ax = flux_points.plot(
    energy_power=2, flux_unit="erg-1 cm-2 s-1", color="darkorange"
)
flux_points.to_sed_type("e2dnde").plot_likelihood(ax=ax)
```

```python
flux_points_dataset = FluxPointsDataset(
    data=flux_points, model=model_best_stacked
)
```

```python
plt.figure(figsize=(8, 6))
flux_points_dataset.peek();
```

```python
model_best_joint.parameters.to_table()
```

```python
model_best_stacked.parameters.to_table()
```

```python
plot_kwargs = {
    "energy_range": [0.1, 30] * u.TeV,
    "energy_power": 2,
    "flux_unit": "erg-1 cm-2 s-1",
}

# plot stacked model
model_best_stacked.plot(**plot_kwargs, label="Stacked analysis result")
model_best_stacked.plot_error(**plot_kwargs)

# plot joint model
model_best_joint.plot(**plot_kwargs, label="Joint analysis result", ls="--")
model_best_joint.plot_error(**plot_kwargs)

#create_crab_spectral_model().plot(**plot_kwargs, label="Crab reference")
plt.legend()
```

# On essaie avec un modèle de loi de puissance avec coupure

%%latex

$\phi(E) = \phi_0 \cdot \left(\frac{E}{E_0}\right)^{-\Gamma} \exp(-\lambda E)$

```python
from gammapy.modeling.models import ExpCutoffPowerLawSpectralModel

model2 = ExpCutoffPowerLawSpectralModel(
     index=2, amplitude=1e-11 * u.Unit("cm-2 s-1 TeV-1"), reference=1 * u.TeV, lambda_ = 1 / u.TeV
)

datasets_joint2 = extraction.spectrum_observations
datasets_stacked2 = Datasets(datasets_joint2).stack_reduce()
```

```python
for dataset in datasets_joint2:
    dataset.model = model2

fit_joint2 = Fit(datasets_joint2)
result_joint2 = fit_joint2.run()

# we make a copy here to compare it later
model_best_joint2 = model2.copy()
model_best_joint2.parameters.covariance = result_joint2.parameters.covariance
```

```python
model_best_joint2.parameters.to_table()
```

```python
datasets_stacked2.model = model2
stacked_fit2 = Fit([datasets_stacked2])
result_stacked2 = stacked_fit2.run()

# make a copy to compare later
model_best_stacked2 = model2.copy()
model_best_stacked2.parameters.covariance = result_stacked2.parameters.covariance
```

```python
model_best_stacked2.parameters.to_table()
```

```python
plt.figure(figsize=(8, 6))
ax_spectrum, ax_residual = datasets_joint2[0].plot_fit()
#ax_spectrum.set_ylim(0, 25)
```

```python
plt.figure(figsize=(8, 6))
ax_spectrum, ax_residual = datasets_stacked2.plot_fit()
#ax_spectrum.set_ylim(0, 25)
```

```python
fpe = FluxPointsEstimator(datasets=datasets_joint2, e_edges=e_edges)
flux_points = fpe.run()

flux_points_dataset = FluxPointsDataset(
    data=flux_points, model=model_best_joint2
)
```

```python
plt.figure(figsize=(8, 6))
flux_points_dataset.peek();
```

```python
fpe = FluxPointsEstimator(datasets=datasets_stacked2, e_edges=e_edges)
flux_points = fpe.run()

flux_points_dataset2 = FluxPointsDataset(
    data=flux_points, model=model_best_stacked2
)
```

```python
plt.figure(figsize=(8, 6))
flux_points_dataset2.peek();
```

```python
plot_kwargs = {
    "energy_range": [0.1, 30] * u.TeV,
    "energy_power": 2,
    "flux_unit": "erg-1 cm-2 s-1",
}

# plot stacked model
model_best_stacked2.plot(**plot_kwargs, label="Stacked analysis result")
model_best_stacked2.plot_error(**plot_kwargs)

# plot joint model
model_best_joint2.plot(**plot_kwargs, label="Joint analysis result", ls="--")
model_best_joint2.plot_error(**plot_kwargs)

#create_crab_spectral_model().plot(**plot_kwargs, label="Crab reference")
plt.legend()
```

## Autre modèle

%%latex

$\phi(E) = \phi_0 \cdot \left(\frac{E}{E_0}\right)^{-\Gamma} \exp \left( \frac{E_0 - E}{E_{C}} \right)$

```python
from gammapy.modeling.models import ExpCutoffPowerLaw3FGLSpectralModel

model3 = ExpCutoffPowerLaw3FGLSpectralModel(
    index=2.0, amplitude=2e-12 * u.Unit("cm-2 s-1 TeV-1"), reference=1 *u.Unit("TeV"), ecut=10 * u.Unit("TeV")
)


datasets_joint3 = extraction.spectrum_observations
datasets_stacked3 = Datasets(datasets_joint3).stack_reduce()

datasets_stacked3.model = model3
stacked_fit3 = Fit([datasets_stacked3])
result_stacked3 = stacked_fit3.run()

# make a copy to compare later
model_best_stacked3 = model3.copy()
model_best_stacked3.parameters.covariance = result_stacked3.parameters.covariance
```

```python
model_best_stacked3.parameters.to_table()
```

```python
plt.figure(figsize=(8, 6))
ax_spectrum, ax_residual = datasets_stacked3.plot_fit()
#ax_spectrum.set_ylim(0, 25)
```

```python
flux_points_dataset3 = FluxPointsDataset(
    data=flux_points, model=model_best_stacked3
)

plt.figure(figsize=(8, 6))
flux_points_dataset3.peek();
```
